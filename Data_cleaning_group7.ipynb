```python
import json
import pandas as pd
import numpy as np

# Specify the path for the input JSON file
json_file_path = 'Cell_Phones_and_Accessories_5.json'

# Create an empty list to store JSON objects
json_data6 = []
# Read JSON data from the file
with open(json_file_path, 'r') as json_file:
    for line in json_file:
        try:
            data = json.loads(line)
            json_data6.append(data)
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON: {e}")
# Create a DataFrame from the list of JSON objects
cell_df = pd.DataFrame(json_data6)
cell_df.head()
```

```python
cell_df.info()
```

```python
# finding the percentage of missing values
total_cells = np.product(cell_df.shape)
total_missing = cell_df.isnull().sum().sum()
percent_missing = (total_missing/total_cells) * 100
print("Percentage of missing values in the dataset: {:.2f}%".format(percent_missing))
```

```python
#replace missing values in Vote with 0
cell_df['vote'] = cell_df['vote'].fillna(0)
cell_df.head()
```

```python
#drop unwanted columns
#We manually identified some features which are not useful for prediction and dropping them.
Features_to_drop = ['reviewText','style','image']
cell_df.drop(labels=Features_to_drop, axis=1, inplace=True)

cell_df.head()
```

```python
unique_styles = cell_df['overall'].unique()
# Print the unique values
print(unique_styles)
```

```python
cell_df['reviewTime'] = pd.to_datetime(cell_df['reviewTime'], format='%m %d, %Y')
cell_df.head()
```

```python
cell_df = cell_df.dropna()
# Drop null value& duplicate rows based on all columns
cell_df.drop_duplicates(inplace=True)
```

```python
cell_df.shape
```

```python
# drop rows with no reviewerName
cell_df.dropna(subset=['reviewerName'], inplace=True)
# Assuming your DataFrame is named df
cell_df = cell_df.dropna(subset=['reviewerName'])
```

```python
cell_df.info()
```

```python
# finding the percentage of missing values
total_cells = np.product(cell_df.shape)
total_missing = cell_df.isnull().sum().sum()
percent_missing = (total_missing/total_cells) * 100
print("Percentage of missing values in the dataset: {:.2f}%".format(percent_missing))
```

```python
cell_df.to_csv('Cell_and_Accessories_clean.csv', index=False)
```
